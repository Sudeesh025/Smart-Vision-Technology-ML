```python
import easyocr
import cv2
import matplotlib.pyplot as plt
# Load the image
img = cv2.imread("C:/users/sudee/Downloads/OIP (1).jpg")

# Initialize the OCR reader
reader = easyocr.Reader(['en'], gpu=False)

# Read text from the image
text = reader.readtext(img)

# Print the results
print(text)
```

```python
for t in text:
    bbox,text_, score = t
    print(t)
    cv2.rectangle(img , bbox[0] , bbox[2] , (0, 255 , 0) , 5)
    cv2.putText(img , text_,bbox[0] , cv2.FONT_HERSHEY_COMPLEX ,1,(255,0,0),2)
```
